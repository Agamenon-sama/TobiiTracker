## Setup

Import necessary dependencies

```python
import matplotlib.pyplot as plt
import presentation
```

Use the presentation script to parse the .ppo files outputed by the post processing software

```python
persons = presentation.parse()
```

The parse function assumes you have an `output/` directory that contains the correctly formatted output of the post processing software.

## Usage

The parse function returns a list `[]` which contains Person data.
A person is an object that has 2 properties:
- "name" the name of the person which is the name of the folder containing his output
- "areas" a list of dictionaries `{}` that contain the following values
    - "name" the name of the file from which aois are extracted. Each file contain the AOIs of particular test.
    - "aois" a list of dictionaries which contain the name of the aoi as key and it's value as a value

```python
# example
personName = persons[0].name
fileName = persons[0].areas[0]["name"]
aoiDict = persons[0].areas[0]["aois"][0]
print(f"{personName}\n{fileName}: {aoiDict}")
```

```python
aois = presentation.interestingAois(persons, "1_aois")
            
plt.bar(aois.keys(), aois.values())

plt.title("Most interesting area in the neptun front page")
plt.xlabel("AOI name")
plt.ylabel("Dwell time average")

plt.grid(axis='y')
```
